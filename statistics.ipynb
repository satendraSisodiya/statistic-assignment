```python
# Q1: Generate a list of 100 integers containing values between 90 to 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i): Write a Python function to calculate the mean of a given list of numbers.
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

mean_value = calculate_mean(int_list)

#  Create a function to find the median of a list of numbers.
def calculate_median(numbers):
    numbers.sort()
    mid = len(numbers) // 2
    if len(numbers) % 2 == 0:
        return (numbers[mid - 1] + numbers[mid]) / 2
    else:
        return numbers[mid]

median_value = calculate_median(int_list)


# Q(ii): Develop a program to compute the mode of a list of integers.
def calculate_mode(numbers):
    return statistics.mode(numbers)

mode_value = calculate_mode(int_list)


# Q(iii): Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
def calculate_weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

weights = [random.randint(1, 5) for _ in range(len(int_list))]
weighted_mean_value = calculate_weighted_mean(int_list, weights)


# Q(iv): Write a Python function to find the geometric mean of a list of positive numbers.
def calculate_geometric_mean(numbers):
    return prod(numbers) ** (1 / len(numbers))

geometric_mean_value = calculate_geometric_mean([n for n in int_list if n > 0])


# Q(v): Create a program to calculate the harmonic mean of a list of values.
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / n for n in numbers if n != 0)

harmonic_mean_value = calculate_harmonic_mean(int_list)

# Q(vi): Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

midrange_value = calculate_midrange(int_list)


# Q(vii): Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.
def calculate_trimmed_mean(numbers, trim_percent):
    trim_count = int(len(numbers) * (trim_percent / 100))
    trimmed_numbers = sorted(numbers)[trim_count:-trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)

trimmed_mean_value = calculate_trimmed_mean(int_list, 10)  # Example: 10% trim


# Print results
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Weighted Mean:", weighted_mean_value)
print("Geometric Mean:", geometric_mean_value)
print("Harmonic Mean:", harmonic_mean_value)
print("Midrange:", midrange_value)
print("Trimmed Mean (10% trim):", trimmed_mean_value)
```

```python
# Q3: Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage
values = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
drv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", drv.expected_value())
print("Variance:", drv.variance())
```

```python
# Q5: Create a Python function to generate random samples from a given probability distribution and calculate their mean and variance.
def random_samples_stats(distribution, params, size=1000):
    if distribution == "binomial":
        n, p = params
        samples = binom.rvs(n=n, p=p, size=size)
    elif distribution == "poisson":
        mu = params[0]
        samples = poisson.rvs(mu=mu, size=size)
    else:
        raise ValueError("Unsupported distribution")

    return np.mean(samples), np.var(samples, ddof=1)

# Example usage for binomial and Poisson distributions
binomial_mean, binomial_variance = random_samples_stats("binomial", (10, 0.5))
print("\nBinomial Distribution - Mean:", binomial_mean, "Variance:", binomial_variance)

poisson_mean, poisson_variance = random_samples_stats("poisson", (3,))
print("Poisson Distribution - Mean:", poisson_mean, "Variance:", poisson_variance)
```

```python
# Q6: Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute their mean, variance, and standard deviation.
def gaussian_distribution_stats(mu=0, sigma=1, size=1000):
    samples = norm.rvs(loc=mu, scale=sigma, size=size)
    mean = np.mean(samples)
    variance = np.var(samples, ddof=1)
    std_dev = np.sqrt(variance)
    return mean, variance, std_dev

# Example usage
gaussian_mean, gaussian_variance, gaussian_std_dev = gaussian_distribution_stats(mu=5, sigma=2)
print("\nGaussian Distribution - Mean:", gaussian_mean)
print("Gaussian Distribution - Variance:", gaussian_variance)
print("Gaussian Distribution - Standard Deviation:", gaussian_std_dev)
```

```python
Q11:A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors Qwho make a purchase). They collect data from the old and new layouts to compare.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

successes = [old_layout.sum(), new_layout.sum()]
totals = [len(old_layout), len(new_layout)]

stat, p_value = proportions_ztest(successes, totals)

print("Z-Statistic:", stat)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout has a statistically significant impact.")
else:
    print("Fail to reject the null hypothesis: The new layout does not have a statistically significant impact.")
```

```python
# Q 14: A customer service department claims that their average response time is less than 5 minutes. A sample of recent customer interactions was taken, and the response times were recorded.

import numpy as np
from statsmodels.stats.weightstats import ztest

response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Null Hypothesis (H0): The mean response time is 5 minutes.
# Alternative Hypothesis (Ha): The mean response time is less than 5 minutes.
stat, p_value = ztest(response_times, value=5, alternative='smaller')

print("Z-Statistic:", stat)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average response time is less than 5 minutes.")
else:
    print("Fail to reject the null hypothesis: There is no evidence that the average response time is less than 5 minutes.")
```

```python
# Q 15: A company is testing two different website layouts to see which one leads to higher click-through rates.
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
# freedom, and p-value.

import numpy as np
from scipy import stats

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

def ab_test_analysis(layout_a, layout_b):

    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)

    std_a = np.std(layout_a, ddof=1)
    std_b = np.std(layout_b, ddof=1)

    n_a = len(layout_a)
    n_b = len(layout_b)

    pooled_std = np.sqrt(((n_a - 1) * std_a**2 + (n_b - 1) * std_b**2) / (n_a + n_b - 2))

    t_stat = (mean_a - mean_b) / (pooled_std * np.sqrt(1/n_a + 1/n_b))

    degrees_of_freedom = n_a + n_b - 2

    p_value = stats.t.sf(np.abs(t_stat), degrees_of_freedom) * 2

    return t_stat, degrees_of_freedom, p_value

t_stat, degrees_of_freedom, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the click-through rates of Layout A and Layout B.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the click-through rates of Layout A and Layout B.")
```

```python
# Q 16: A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in
# reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the t-statistic and p-value for the treatment effect.

import numpy as np
from scipy import stats

existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

def drug_effect_analysis(existing_drug, new_drug):

    mean_existing = np.mean(existing_drug)
    mean_new = np.mean(new_drug)

    std_existing = np.std(existing_drug, ddof=1)
    std_new = np.std(new_drug, ddof=1)

    n_existing = len(existing_drug)
    n_new = len(new_drug)

    pooled_std = np.sqrt(((n_existing - 1) * std_existing**2 + (n_new - 1) * std_new**2) / (n_existing + n_new - 2))

    t_stat = (mean_existing - mean_new) / (pooled_std * np.sqrt(1/n_existing + 1/n_new))

    degrees_of_freedom = n_existing + n_new - 2

    p_value = stats.t.sf(np.abs(t_stat), degrees_of_freedom) * 2

    return t_stat, degrees_of_freedom, p_value

t_stat, degrees_of_freedom, p_value = drug_effect_analysis(existing_drug_levels, new_drug_levels)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new drug is more effective than the existing drug in reducing cholesterol levels.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the new drug and the existing drug.")
```

```python
# Q 18: An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
# a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
# significant difference between the average salaries of male and female employees.

import numpy as np
from scipy import stats

np.random.seed(0)

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def salary_gap_analysis(male_salaries, female_salaries):
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)

    std_male = np.std(male_salaries, ddof=1)
    std_female = np.std(female_salaries, ddof=1)

    n_male = len(male_salaries)
    n_female = len(female_salaries)

    pooled_std = np.sqrt(((n_male - 1) * std_male**2 + (n_female - 1) * std_female**2) / (n_male + n_female - 2))

    t_stat = (mean_male - mean_female) / (pooled_std * np.sqrt(1/n_male + 1/n_female))

    degrees_of_freedom = n_male + n_female - 2

    p_value = stats.t.sf(np.abs(t_stat), degrees_of_freedom) * 2

    return t_stat, degrees_of_freedom, p_value

t_stat, degrees_of_freedom, p_value = salary_gap_analysis(male_salaries, female_salaries)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant salary gap between male and female employees.")
else:
    print("Fail to reject the null hypothesis: There is no significant salary gap between male and female employees.")
```

```python
# Q 19: A manufacturer produces two different versions of a product and wants to compare their quality scores.
# Create a Python function to analyze quality assessment data, calculate the t-statistic, and decide
# whether there's a significant difference in quality between the two versions.

import numpy as np
from scipy import stats

version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

def quality_comparison(version1_scores, version2_scores):
    mean_v1 = np.mean(version1_scores)
    mean_v2 = np.mean(version2_scores)

    std_v1 = np.std(version1_scores, ddof=1)
    std_v2 = np.std(version2_scores, ddof=1)

    n_v1 = len(version1_scores)
    n_v2 = len(version2_scores)

    pooled_std = np.sqrt(((n_v1 - 1) * std_v1**2 + (n_v2 - 1) * std_v2**2) / (n_v1 + n_v2 - 2))

    t_stat = (mean_v1 - mean_v2) / (pooled_std * np.sqrt(1/n_v1 + 1/n_v2))

    degrees_of_freedom = n_v1 + n_v2 - 2

    p_value = stats.t.sf(np.abs(t_stat), degrees_of_freedom) * 2

    return t_stat, degrees_of_freedom, p_value

t_stat, degrees_of_freedom, p_value = quality_comparison(version1_scores, version2_scores)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in quality between the two versions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in quality between the two versions.")
```

```python
# Q 20:  A restaurant chain collects customer satisfaction scores for two different branches. Write a program to
# analyze the scores, calculate the t-statistic, and determine if there's a statistically significant difference in
# customer satisfaction between the branches.

import numpy as np
from scipy import stats

branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

def satisfaction_comparison(branch_a_scores, branch_b_scores):
    mean_a = np.mean(branch_a_scores)
    mean_b = np.mean(branch_b_scores)

    std_a = np.std(branch_a_scores, ddof=1)
    std_b = np.std(branch_b_scores, ddof=1)

    n_a = len(branch_a_scores)
    n_b = len(branch_b_scores)

    pooled_std = np.sqrt(((n_a - 1) * std_a**2 + (n_b - 1) * std_b**2) / (n_a + n_b - 2))

    t_stat = (mean_a - mean_b) / (pooled_std * np.sqrt(1/n_a + 1/n_b))

    degrees_of_freedom = n_a + n_b - 2

    p_value = stats.t.sf(np.abs(t_stat), degrees_of_freedom) * 2

    return t_stat, degrees_of_freedom, p_value

t_stat, degrees_of_freedom, p_value = satisfaction_comparison(branch_a_scores, branch_b_scores)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in customer satisfaction between the two branches.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in customer satisfaction between the two branches.")
```

```python
# Q 21: A political analyst wants to determine if there is a significant association between age groups and voter
# preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
# them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
# there is a significant association between age groups and voter preferences.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

np.random.seed(0)

age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)


data = pd.crosstab(age_groups, voter_preferences)

chi2_stat, p_value, dof, expected = chi2_contingency(data)

print("Chi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_value)
print("Expected Frequencies:\n", expected)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")
```

```python
# Q 22: A company conducted a customer satisfaction survey to determine if there is a significant relationship
# between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are
# located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a Chi
# Square test to determine if there is a significant relationship between product satisfaction levels and
# customer regions.

import numpy as np
import scipy.stats as stats

data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

print("Chi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_value)
print("Expected Frequencies:\n", expected)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant relationship between product satisfaction levels and customer regions.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between product satisfaction levels and customer regions.")
```

```python
# Q 23: A company implemented an employee training program to improve job performance (Effective, Neutral,
# Ineffective). After the training, they collected data from a sample of employees and classified them based
# on their job performance before and after the training. Perform a Chi-Square test to determine if there is a
# significant difference between job performance levels before and after the training.

import numpy as np
import scipy.stats as stats

data = np.array([[50, 30, 20], [30, 40, 30], [20, 30, 40]])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

print("Chi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_value)
print("Expected Frequencies:\n", expected)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in job performance levels before and after the training.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in job performance levels before and after the training.")
```

```python
# Q 24: A company produces three different versions of a product: Standard, Premium, and Deluxe. The
# company wants to determine if there is a significant difference in customer satisfaction scores among the
# three product versions. They conducted a survey and collected customer satisfaction scores for each
# version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
# difference in customer satisfaction scores.

import numpy as np
import scipy.stats as stats

standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

f_statistic, p_value = stats.f_oneway(standard_scores, premium_scores, deluxe_scores)

print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in customer satisfaction scores among the product versions.")
```
